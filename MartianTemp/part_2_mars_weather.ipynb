```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
from pprint import pprint
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
html=browser.html
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract all rows of data
table = soup.find('table', class_='table')
rows = table.find_all('tr')
rows
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Heres an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
mars_temp_list = []
# Loop through the scraped data to create a list of rows
for row in rows:
    cells = row.find_all(['th','td'])
    cell_data = [cell.get_text(strip=True) for cell in cells]
    mars_temp_list.append(cell_data)

pprint(mars_temp_list)
```

```python
# Use the first row as the column names and the rest as data
columns = mars_temp_list[0]  # The first row contains the column names
data = mars_temp_list[1:]    # The rest of the rows contain the data
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
mars_df = pd.DataFrame(data, columns=columns)
```

```python
# Confirm DataFrame was created successfully
mars_df.head()
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
mars_df.dtypes
```

```python
# Change data types for data analysis
mars_df['terrestrial_date'] = pd.to_datetime(mars_df['terrestrial_date'])

# Convert other columns to their respective types
mars_df = mars_df.astype({
    'sol': 'int32',
    'ls': 'int32',
    'month': 'int32',
    'min_temp': 'float64',
    'pressure': 'float64'
})
```

```python
# Confirm type changes were successful by examining data types again
mars_df.dtypes
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
month_count = mars_df['month'].nunique()
month_count
```

```python
# 2. How many sols (Martian days) worth of data are there?
martian_days = mars_df['sol'].nunique()
martian_days
```

```python
# 3. What is the average minimum temperature by month?
avg_min_temp_month = mars_df.groupby('month')['min_temp'].mean()

# What are the coldest and the warmest months on Mars?
coldest_avg_temp = avg_min_temp_month.idxmin() #coldest month 
print(f' The average coldest month is month {coldest_avg_temp}') 

warmest_avg_temp = avg_min_temp_month.idxmax() #warmest month
print(f' The average warmest month is month {warmest_avg_temp}')

avg_min_temp_month
```

```python
# Plot the average minimum temperature by month
avg_min_temp_month.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Temperature in Celsius')
plt.title('Average Minimum Temperature By Month')
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location by sorting the previous graph
avg_min_temp_month_sorted = avg_min_temp_month.sort_values()
avg_min_temp_month_sorted.plot(kind = 'bar')
plt.xlabel('Month')
plt.ylabel('Temperature in Celsius')
plt.title('Average Minimum Temperature By Month')
plt.show()
```

```python
# 4. What is the average pressure by month?
avg_press_per_month = mars_df.groupby('month')['pressure'].mean()

#Which months have the lowest and the highest atmospheric pressure on Mars?
lowest_avg_pressure = avg_press_per_month.idxmin()
print(f' The average lowest pressure month is month {lowest_avg_pressure}') 

highest_avg_pressure = avg_press_per_month.idxmax()
print(f' The average highest pressure month is month {highest_avg_pressure}')

avg_press_per_month
```

```python
# Plot the average pressure by month
avg_press_per_month.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Atomspheric Pressure')
plt.title('Average Pressure by Month')
plt.show()
```

```python
# Identify the lowest and highest pressure months in Curiosity's location by sorting the previous graph
avg_press_per_month_sorted = avg_press_per_month.sort_values()
avg_press_per_month_sorted.plot(kind = 'bar')
plt.xlabel('Month')
plt.ylabel('Atomspheric Pressure')
plt.title('Average Pressure by Month')
plt.show()
```

```python
# 5. How many terrestrial (Earth) days are there in a Martian year?
x_axis = range(martian_days) #create a variable for the amount of martian days
print(f"The amount of martian days that exist are: {martian_days}")

# Visually estimate the result by plotting the daily minimum temperature of each observation in the data set.
plt.plot(x_axis, mars_df['min_temp'])
plt.xlabel('Number of Terrestrial Date')
plt.ylabel('Minimum Temperature')
plt.title('Minimum Temperature over Time')
plt.grid()
plt.show()
```

#### Minimum Temperature
Looking at the bar graph called 'Average Minimum Temperature by Month' it can be seen that the average coldest month is month 3 and the average warmest month is 8.


#### Atmospheric Pressure
For Atmospheric Pressure, a bar graph was plotted under the title 'Average Pressure by Month' to visualize which months had the highest average pressure and which months had the lowest pressure month. What was found is that the avergae lowest pressure month was month 6 and the average highest pressure month was month 9. 


#### Year Length

When looking at the graph "Minimum Temperature over Time" we can roughly estimate how long a martian year is. If we go based on a seasonal change which is expressed on Earth as a gradual change in temperature then it can be estimated that Martian years are about 600-700 days long. The graph is sinuisoidal in nature, so by taking a look at the beginning of the graph to the beginning of the next phase that change occurs around 600 - 700 days long.
